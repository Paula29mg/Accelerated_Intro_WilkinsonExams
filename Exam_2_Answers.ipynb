```python
def first_column(file,lista):
    gfile = open(file,"r")
    n = 0
    for line in gfile:
        if n == 0:
            n += 1
        else:
            line = line.split('\t')[0]
            lista.append(line)

def comp_list (lista1,lista2):
    n = 0
    for x in lista1:
        if x == lista2[n]:
            n += 1
            print("true")
        else:
            print(lista1[n] + "is not here")

A = []
first_column("Germplasm.tsv",A)
B = []
first_column("LocusGene.tsv",B)

comp_list(A,B)
```

## PROBLEM 3

```python
import pymysql.cursors
import csv
connection = pymysql.connect(host='localhost',
                            user='root',
                            password='root',
                            db='examdatabase',
                            charset='utf8mb4', 
                            cursorclass=pymysql.cursors.DictCursor)
connection.autocommit = False

 

try:
    with connection.cursor() as cursor:
        with open('Germplasm.tsv') as csvfile:
            Germplasm = csv.DictReader(csvfile, delimiter = '\t')
            for infogermplasm in Germplasm:
                print(infogermplasm)
                sql = """INSERT INTO Germplasm (Locus, Germplasm, Phenotype, Pubmed) values ('{}', '{}', '{}', '{}')""".format(infogermplasm["Locus"], infogermplasm["germplasm"], infogermplasm["phenotype"], infogermplasm["pubmed"])
                cursor.execute(sql)
        connection.commit() 
finally: 
    connection.close()
```

```python
import pymysql.cursors
import csv
connection = pymysql.connect(host='localhost',
                            user='root',
                            password='root',
                            db='examdatabase',
                            charset='utf8mb4', 
                            cursorclass=pymysql.cursors.DictCursor)
connection.autocommit = False

 

try:
    with connection.cursor() as cursor:
        with open('LocusGene.tsv') as csvfile:
            LocusGene = csv.DictReader(csvfile, delimiter = '\t')
            for infolocus in LocusGene:
                print(infolocus)
                sql = """INSERT INTO LocusGene (Locus, Gene, Proteinlength) values ('{}', '{}', '{}')""".format(infolocus["Locus"], infolocus["Gene"], infolocus["ProteinLength"])
                cursor.execute(sql)
        connection.commit() 
finally: 
    connection.close()
```
